```python
import numpy as np
from gensim.models.doc2vec import Doc2Vec
from gensim.models.word2vec import Word2Vec
from owlready2 import *
# Import from Project
import Competence_Assessment
import Compute_Learning_Path
import Document_Preprocessing_and_Tagging
import DSPP_IO
import Ngram_Management
import Similarity
import TensorFlow_Operations
```

```python
dspp = 'DSP04'
fileName = 'Mapping_DSP.txt'
path = './data/competences.jsonl'
```

```python
competencies_json = Document_Preprocessing_and_Tagging.readFile(path)
corpus = Document_Preprocessing_and_Tagging.splitAndTag(competencies_json)
```

```python
# model.window = 5 by default
model = Doc2Vec(dm=0, vector_size=100, min_count=40, epochs=30) 
model.build_vocab(corpus)
```

```python
ng_list = Ngram_Management.getNGramFromDocuments(competencies_json, 2)
bigrams = Ngram_Management.computeCommonNgrams(ng_list, 40)
```

```python
model1 = Word2Vec(min_count = 1, size = 100, window = 5) 
model1.build_vocab(bigrams[1])
model.wv.vocab.update(model1.wv.vocab)
```

```python
#PraparaRimozione
coordinates = np.concatenate([model.wv.vectors, model1.wv.vectors])
```

```python
for item in bigrams[0]:
    entry = item.split()
    coordinates = TensorFlow_Operations.removeKey(entry[0], model, coordinates)
    coordinates = TensorFlow_Operations.removeKey(entry[1], model, coordinates)
    coordinates = TensorFlow_Operations.removeKey(entry[0]+entry[1], model, coordinates)
```

```python
TensorFlow_Operations.tensorBoardOutput(model, coordinates)
```

```python
%time model.train(corpus, total_examples=model.corpus_count, epochs=model.epochs)
```
